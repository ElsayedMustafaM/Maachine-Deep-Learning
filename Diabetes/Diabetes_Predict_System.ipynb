```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier,AdaBoostClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import SGDClassifier
from sklearn.gaussian_process import GaussianProcessClassifier
from sklearn.preprocessing import Imputer
from sklearn.svm import SVC
from sklearn.svm import LinearSVC
from sklearn.neighbors import KNeighborsClassifier
from tkinter import *
from tkinter import ttk
from tkinter import messagebox

    
```

```python

data=pd.read_csv('diabetes.csv')
names=['Pregnancies','Glucose','BloodPressure','SkinThickness','Insulin','BMI','DiabetesPedigreeFunction','Age']
features=data[list(names)].values
label=data['Outcome'].values

imputer =Imputer(missing_values='NaN',strategy='mean',axis=0)
features=imputer.fit_transform(features)

xtrain,xtest,ytrain,ytest=train_test_split(features,label,test_size=0.30)

feut=[]



def display(y):
    master1=Tk()
    master1.title("Diabetes Predict System")
    master1.geometry('300x220+500+200')
    imgg=PhotoImage(file='sad.png')
    imgg1=PhotoImage(file='happy.png')
    res=imgg1.subsample(2,3)

    
    if (y==1):
        label1 = ttk.Label(master1,text='Congratulations you are fine', font=50)
        master1.configure(background='white')
        label1.configure(background='white', foreground='black')
        label1.pack()
        label1.place(x=20, y=20)
        label2 = ttk.Label(image=res)
        label2.place(x=0, y=50)
        
        
        
    else:
        master1.configure(background='black')
        label1 = ttk.Label(master1,text='Unfortunately you are sick', font=50)
        label1.configure(background='black', foreground='white')
        label1.pack()
        label1.place(x=20, y=20)
        label2 = ttk.Label(image=imgg)
        label2.place(x=0, y=50)
        

    master1.mainloop()

def Svc():
    subm()
    clf= SVC()
    clf=clf.fit(xtrain,ytrain)
    ypredict=clf.predict(np.array(feut).reshape(1,-1))
    master.destroy()
    display(ypredict)
    
def Mlplassifier():
    subm()
    clf= MLPClassifier()
    clf=clf.fit(xtrain,ytrain)
    ypredict=clf.predict(np.array(feut).reshape(1,-1))
    master.destroy()
    display(ypredict)

def KNeighbors():
    subm()
    clf= KNeighborsClassifier()
    clf=clf.fit(xtrain,ytrain)
    ypredict=clf.predict(np.array(feut).reshape(1,-1))
    master.destroy()
    display(ypredict)

def RandomForest():
    subm()
    clf=RandomForestClassifier()
    clf=clf.fit(xtrain,ytrain)
    ypredict=clf.predict(np.array(feut).reshape(1,-1))
    master.destroy()
    display(ypredict)
    

def LinearSvc():
    subm()
    clf= LinearSVC()
    clf=clf.fit(xtrain,ytrain)
    ypredict=clf.predict(np.array(feut).reshape(1,-1))
    master.destroy()
    display(ypredict)

def GaussianprocessClassifier():
    subm()
    clf= GaussianProcessClassifier()
    clf=clf.fit(xtrain,ytrain)
    ypredict=clf.predict(np.array(feut).reshape(1,-1))
    master.destroy()
    display(ypredict)
    
    
def AdaboostClassifier():
    subm()
    clf= AdaBoostClassifier()
    clf=clf.fit(xtrain,ytrain)
    ypredict=clf.predict(np.array(feut).reshape(1,-1))
    master.destroy()
    display(ypredict)
    
    
def naive_bayes():
    subm()
    clf= GaussianNB()
    clf=clf.fit(xtrain,ytrain)
    ypredict=clf.predict(np.array(feut).reshape(1,-1))
    master.destroy()
    display(ypredict)
    
def DecisionTree():
    subm()
    clf= DecisionTreeClassifier()
    clf=clf.fit(xtrain,ytrain)
    ypredict=clf.predict(np.array(feut).reshape(1,-1))
    master.destroy()
    display(ypredict)  
    
```
