```python
import pandas as pd
import numpy as np
from sklearn import preprocessing,cross_validation
from sklearn.neighbors import KNeighborsClassifier
from sklearn.multiclass import OneVsRestClassifier
from sklearn.preprocessing import Imputer
from tkinter import *
from tkinter import ttk

data=pd.read_csv('diabetes.csv')
names=['Pregnancies','Glucose','BloodPressure','SkinThickness','Insulin','BMI','DiabetesPedigreeFunction','Age']
features=data[list(names)].values
label=data['Outcome'].values

imputer =Imputer(missing_values='NaN',strategy='mean',axis=0)
features=imputer.fit_transform(features)

xtrain,xtest,ytrain,ytest=cross_validation.train_test_split(features,label,test_size=0.30)

res=None
feut=[]
master=Tk()

master.configure(background='#180145224')
master.title("Diabetes Predict System")
master.geometry('1000x540+170+100')

 #labels
feature1=ttk.Label(None,text="Age : ",font=50)
feature2=ttk.Label(None,text="BMI : ",font=40)
feature3=ttk.Label(None,text="Insulin : ",font=40)
feature4=ttk.Label(None,text="Glucose : ",font=40)
feature5=ttk.Label(None,text="Pregnancies : ",font=40)
feature6=ttk.Label(None,text="BloodPressure : ",font=40)
feature7=ttk.Label(None,text="SkinThickness : ",font=40)
feature8=ttk.Label(None,text="DiabetesPedigree : ",font=40)

feature1.pack()
feature2.pack()
feature3.pack()
feature4.pack()
feature5.pack()
feature6.pack()
feature7.pack()
feature8.pack()
#feature2.configure(background='#038046117',foreground='#902739')
feature1.configure(background='#180145224',foreground='white')
feature2.configure(background='#180145224',foreground='white')
feature3.configure(background='#180145224',foreground='white')
feature4.configure(background='#180145224',foreground='white')
feature5.configure(background='#180145224',foreground='white')
feature6.configure(background='#180145224',foreground='white')
feature7.configure(background='#180145224',foreground='white')
feature8.configure(background='#180145224',foreground='white')
feature1.place(x=40,y=100)
feature2.place(x=40,y=150)
feature3.place(x=40,y=200)
feature4.place(x=40,y=250)
feature5.place(x=40,y=300)
feature6.place(x=40,y=350)
feature7.place(x=40,y=400)
feature8.place(x=40,y=450)

#Entries
entry1=ttk.Entry()
entry2=ttk.Entry()
entry3=ttk.Entry()
entry4=ttk.Entry()
entry5=ttk.Entry()
entry6=ttk.Entry()
entry7=ttk.Entry()
entry8=ttk.Entry()
#entry1.configure(foreground='#983044')
entry1.pack()
entry2.pack()
entry3.pack()
entry4.pack()
entry5.pack()
entry6.pack()
entry7.pack()
entry8.pack()


entry1.place(x=210,y=100,width=150)
entry2.place(x=210,y=150,width=150)
entry3.place(x=210,y=200,width=150)
entry4.place(x=210,y=250,width=150)
entry5.place(x=210,y=300,width=150)
entry6.place(x=210,y=350,width=150)
entry7.place(x=210,y=400,width=150)
entry8.place(x=210,y=450,width=150)

Res=ttk.Label(None,font=50,width=20)
Res.pack()
Res.place(x=700,y=100)





def Result():
    subm()
    clf=OneVsRestClassifier(KNeighborsClassifier(),n_jobs=-1)
    clf=clf.fit(xtrain,ytrain)
    ypredict=clf.predict_proba(np.array(feut).reshape(1,-1))
    res=ypredict[0][0]
    res=res*100
    feut.clear()
    Res.configure(text='{} %'.format (res))
    Res.pack()
    Res.place(x=700,y=100)





    
def subm():
    
    feut.append(float(entry1.get()))
    feut.append(float(entry2.get()))
    feut.append(float(entry3.get()))
    feut.append(float(entry4.get()))
    feut.append(float(entry5.get()))
    feut.append(float(entry6.get()))
    feut.append(float(entry7.get()))
    feut.append(float(entry8.get()))
    entry1.delete(0,END)
    entry2.delete(0,END)
    entry3.delete(0,END)
    entry4.delete(0,END)
    entry5.delete(0,END)
    entry6.delete(0,END)
    entry7.delete(0,END)
    entry8.delete(0,END)




label=ttk.Label(None,text="The percent of disease :  ",font=100)
label.pack()
label.place(x=500,y=100)
label.configure(background='#180145224',foreground='white')


alg1=ttk.Button(text='Result ',width=30,command=Result)

alg1.pack()

alg1.place(x=600,y=200)


master.mainloop()
```
